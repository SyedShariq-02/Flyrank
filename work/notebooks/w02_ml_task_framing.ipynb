## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*
*A proxy binary label indicating whether a page experienced significant performance decay requiring intervention (needs_refresh = 1).

Source of Label: Derived rule based on observed outcomes in historical search performance—specifically defined as pages where recent traffic decreased by more than 20% compared to baseline (clicks_recent < 0.8 * clicks_baseline) while overall search demand remained stable.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# Load dataset to verify target label rule
url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Compute target label proxy
df['needs_refresh'] = (df['trend_direction'] == 'down').astype(int)
print(f"Target distribution (needs_refresh):\n{df['needs_refresh'].value_counts(normalize=True).round(3)}")
```

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Verify metric targets
target_p_at_k = 0.80
k_eval = 500

print(f"Primary Evaluation Metric: Precision@{k_eval}")
print(f"Minimum Acceptable Threshold: {target_p_at_k * 100}% precision in top-{k_eval} recommendations.")
```

## 5. Why ML beats a fixed rule here

ML models can weigh dozens of interacting non-linear signals simultaneously (CTR changes, position shifts, SERP feature volatility, and historical baseline variance) to isolate true content decay from temporary noise.*
