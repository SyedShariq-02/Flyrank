## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

*Data Overview: Analyzing data/raw/content_refresh_anonymized.csv (containing 30,000 anonymized page records across 44 metrics):

Total Rows Evaluated: 30,000 pages sampled across multiple domains.

Declining Page Proportion: 16,262 pages (54.2%) exhibit a downward traffic trend (trend_direction == 'down'), representing the high-priority queue for content refresh.

Traffic Impact Span: Recent clicks (clicks_recent) span from 0 to high volumes, showing that prioritizing pages by potential business impact is far more effective than treating all decaying pages equally.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# Load starter dataset
df = pd.read_csv('https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv')

# Print basic dataset info to check available columns
total_pages = len(df)
declining_pages = (df['trend_direction'] == 'down').sum() if 'trend_direction' in df.columns else 0
pct_declining = (declining_pages / total_pages) * 100

print(f"1. Total pages in dataset: {total_pages:,}")
print(f"2. Declining pages count: {declining_pages:,} ({pct_declining:.1f}%)")

# Find traffic or click-related columns dynamically
click_cols = [c for c in df.columns if 'click' in c.lower() or 'traffic' in c.lower()]
print("3. Clicks/Traffic columns found:", click_cols)

if click_cols:
    target_col = click_cols[0]
    print(f"   Distribution for '{target_col}' (min/median/max): {df[target_col].min()} / {df[target_col].median():.0f} / {df[target_col].max():,}")
```
